# Transforms
The transforms from depth to color for the lowewr camera on the robot appear to be quite good. But the upper camera appears to be mis-aligned. Exploring to transformations JSON file to see if we can find out why.

```python
import pathlib
import json
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline 

directory = pathlib.Path('/home/mjsobrep/Downloads/data/')
with open(directory/'transforms.json', encoding='utf-8') as json_file:
    transforms = json.load(json_file)
```

```python
def plt_trans(trans):
    plt.figure(figsize=(8,5), dpi= 200, facecolor='w', edgecolor='k')
    plt.subplot(311)
    plt.plot(trans[:,0])
    plt.subplot(312)
    plt.plot(trans[:,1])
    plt.subplot(313)
    plt.plot(trans[:,2])
```

```python
trans=(np.array([trans['translation'] for trans in transforms['robot']['upper'].values()]))
plt_trans(trans)
```

```python
rot=(np.array([trans['rotation'] for trans in transforms['robot']['upper'].values()]))
np.any(np.diff(rot, axis=0))
```

Literally no change ever. That is disapointing. I thought these things were being calibrated ocassionally...
